```python
import numpy as np
import matplotlib.pyplot as plt

# Función a minimizar 
def f(x):
    return x**2

# Derivada de la función 
def df(x):
    return 2*x

# Implementación del descenso de gradiente
def descenso_gradiente(tasa_aprendizaje, x_inicial, num_iteraciones):
    x = x_inicial
    valores_x = [x]
    for i in range(num_iteraciones):
        x -= tasa_aprendizaje * df(x)
        valores_x.append(x)
    return x, valores_x

tasa_aprendizaje = 0.1 
x_inicial = 10
num_iteraciones = 50

x_final, valores_x = descenso_gradiente(tasa_aprendizaje, x_inicial, num_iteraciones)

rango_x = np.linspace(-x_inicial, x_inicial, 400)
rango_y = f(rango_x)

plt.figure(figsize=(10, 6))
plt.plot(rango_x, rango_y, label='$f(x) = x^3$')
plt.scatter(valores_x, [f(x) for x in valores_x], color='red')
plt.plot(valores_x, [f(x) for x in valores_x], color='red', linestyle='--')
plt.title('Descenso del Gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
