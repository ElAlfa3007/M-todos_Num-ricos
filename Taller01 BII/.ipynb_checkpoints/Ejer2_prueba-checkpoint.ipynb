```python
import re

def hacer_distintos(A):
    operaciones = 0
    vistos = set()
    
    for i in range(len(A)):
        num = A[i]
        if num in vistos:
            incrementos = 0
            while num + incrementos in vistos:
                incrementos += 1
            decrementos = 0
            while num - decrementos in vistos:
                decrementos += 1
            if incrementos <= decrementos:
                A[i] = num + incrementos
                vistos.add(num + incrementos)
                operaciones += incrementos
            else:
                A[i] = num - decrementos
                vistos.add(num - decrementos)
                operaciones += decrementos
        else:
            vistos.add(num)
    
    return operaciones

def obtener_numeros(cadena):
    return list(map(int, re.findall(r'-?\d+', cadena)))

N = int(input().strip())
entrada_A = input().strip()
A = obtener_numeros(entrada_A)
A = A[:N]

resultado = hacer_distintos(A)
print(resultado)
```

```python
import re
def hacer_distintos(A):
    A.sort()
    operaciones = 0
    vistos = set()
    
    for i in range(len(A)):
        num = A[i]
        if num in vistos:
            incrementos = 0
            decrementos = 0
            new_num = num
            while new_num in vistos:
                new_num += 1
                incrementos += 1
            new_num_decremento = num - 1
            while new_num_decremento in vistos and new_num_decremento > 0:
                new_num_decremento -= 1
                decrementos += 1
            if incrementos <= decrementos:
                A[i] = num + incrementos
                vistos.add(num + incrementos)
                operaciones += incrementos
            else:
                A[i] = new_num_decremento
                vistos.add(new_num_decremento)
                operaciones += decrementos
        else:
            vistos.add(num)
    
    return operaciones

def obtener_numeros(cadena):
    return list(map(int, re.findall(r'\d+', cadena)))

N = int(input().strip())
entrada_A = input().strip()
A = obtener_numeros(entrada_A)
A = A[:N]

resultado = hacer_distintos(A)
print(resultado)
```
