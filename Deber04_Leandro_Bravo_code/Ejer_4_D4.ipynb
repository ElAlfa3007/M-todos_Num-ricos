```python
import numpy as np

import matplotlib.pyplot as plt

def equation1(x:float)->float:
    return ((x**2) - 1)

def equation2(x:float)->float:
    return np.exp(1-x**2)

# Generate x values
x = np.linspace(-5, 5, 200)

# Calculate y values
y1 = equation1(x)
y2 = equation2(x)
# Plot the equation
plt.plot(x, y1, label = '$x^{2}-1$', color='purple')
plt.plot(x, y2, label = '$e^{1-x^{2}}$', color='pink')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $x^{2}-1$ and $e^{1-x^{2}}$')
ax = plt.gca()
ax.set_ylim([-2, 5])
ax.set_xlim([-4, 4])
plt.grid(True)
plt.legend()
# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.savefig('Grafico4a.jpg')
plt.show()
```

```python
def eqBis(x):
    g = (x**2) - 1
    h = np.exp(1-x**2)
    return g-h

# Genera los valores de x
x = np.linspace(-5, 5, 100)
# Calcula los valores de y 
y = eqBis(x)
# Plot the equation
plt.plot(x, y, label = '$g(x)-h(x)$', color = 'blue')
#Etiqueta los ejes
plt.xlabel('x')
plt.ylabel('y')
#añade un título a la gráfica
plt.title('Plot of $f(x)$')
#Configura los límites de los ejes
ax = plt.gca()
ax.set_ylim([-2, 5])
ax.set_xlim([-4, 4])
#añade una cuadrícula a la gráfica
plt.grid(True)
#Añade la leyenda
plt.legend()
# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
#Muestra la gráfica
plt.savefig('GraficoE4.jpg')
plt.show()
```

```python
from typing import Callable
def sign(x: float) -> int:
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
```

```python
def bisection(
    a: float, b: float, *, equation: Callable[[float], float], tol: float, N: int
) -> tuple[float, float, float, int] | None:
    i = 1
    assert a < b, "a not lower than b, the interval is not valid."

    assert (equation(a) * equation(b) < 0), "The function does not change sign over the interval."

    Fa = equation(a)
    p = a # to avoid crash in i == 0.
    for i in range(N):
        p = a + (b - a) / 2
        FP = equation(p)
        if FP == 0 or (b - a) / 2 < tol:
            return p, a, b, i

        if sign(Fa) * sign(FP) > 0:
            a = p
            Fa = FP
        else:
            b = p
    print("Solución no encontrada, número de iteraciones insuficiente para la tolerancia dada.")
```

```python
a = -2
b = 0
tol = 10**(-3)
result = bisection(a=a,b=b,equation=eqBis,tol=tol,N=20)
print("Al utilizar el rango ["+str(a)+","+str(b)+"], en la iteración n°: "+str(result[3])+" se encontró que la raíz de f(x)"+
      " dentro de la precisión de "+format(tol, ".0e")+ " es: "+str(result[0])) 
```
