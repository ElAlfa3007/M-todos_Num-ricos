```python
def equation(x):
    return (x+3)*(x+1)**2*x*(x-1)**3*(x-3)
```

```python
import numpy as np

import matplotlib.pyplot as plt
# Generate x values
x = np.linspace(-10, 10, 1000)

# Calculate y values
y = equation(x)
# Plot the equation
plt.figure()
plt.plot(x, y, label = '$f(x)$', color='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $f(x)$')
ax = plt.gca()
ax.set_ylim([-5, 2])
ax.set_xlim([-4, 4])
plt.grid(True)
plt.legend()
# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.savefig('Grafico5A.jpg')
plt.show()

plt.figure()
plt.plot(x, y, label = '$f(x)$', color='brown')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $f(x)$')
ax = plt.gca()
ax.set_ylim([-700, 50])
ax.set_xlim([-4, 4])
plt.grid(True)
plt.legend()
# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.savefig('GraficoE5.jpg')
plt.show()
```

```python
from typing import Callable
def sign(x: float) -> int:
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
```

```python
def bisection(
    a: float, b: float, *, equation: Callable[[float], float], tol: float, N: int
) -> tuple[float, float, float, int] | None:
    i = 1
    assert a < b, "a not lower than b, the interval is not valid."

    assert (sign(equation(a)) * sign(equation(b)) <= 0), "The function does not change sign over the interval."

    Fa = equation(a)
    p = a # to avoid crash in i == 0.
    for i in range(N):
        p = a + (b - a) / 2
        FP = equation(p)
        if FP == 0 or (b - a) / 2 < tol:
            return p, a, b, i

        if sign(Fa) * sign(FP) > 0:
            a = p
            Fa = FP
        else:
            b = p
    print("Solución no encontrada, número de iteraciones insuficiente para la tolerancia dada.")
```

#### a) [-1.5,2.5]

```python
a = -1.5
b = 2.5
tol = 10**(-5)
result = bisection(a=a,b=b,equation=equation,tol=tol,N=20)
print("Al utilizar el rango ["+str(a)+","+str(b)+"], en la iteración n°: "+str(result[3])+" se encontró que la raíz de f(x)"+
      " dentro de la precisión de "+format(tol, ".0e")+ " es: "+str(result[0])) 
```
