```python
import numpy as np

import matplotlib.pyplot as plt

def equation1(x:float)->float:
    return x

def equation2(x:float)->float:
    return np.tan(x)

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y1 = equation1(x)
y2 = equation2(x)
# Plot the equation
plt.plot(x, y1, label = 'x',color='red')
plt.plot(x, y2, label = 'sin(x)',color='green')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $x$ and $sin(x)$')
ax = plt.gca()
ax.set_ylim([-5, 5])
ax.set_xlim([-5, 5])
plt.grid(True)
plt.legend()
# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.savefig('GraficoE3.jpg')
plt.show()
```

```python
from typing import Callable
def sign(x: float) -> int:
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
```

```python
def bisection(
    a: float, b: float, *, equation: Callable[[float], float], tol: float, N: int
) -> tuple[float, float, float, int] | None:
    i = 1
    assert a < b, "a not lower than b, the interval is not valid."

    assert (equation(a) * equation(b) < 0), "The function does not change sign over the interval."

    Fa = equation(a)
    p = a # to avoid crash in i == 0.
    for i in range(N):
        p = a + (b - a) / 2
        FP = equation(p)
        if FP == 0 or (b - a) / 2 < tol:
            return p, a, b, i

        if sign(Fa) * sign(FP) > 0:
            a = p
            Fa = FP
        else:
            b = p
    print("Solucin no encontrada, nmero de iteraciones insuficiente para la tolerancia dada.")
```
